```python
import getpass
ACCESS_KEY = getpass.getpass("Enter your AWS ACCESS_KEY")
SECRET_KEY = getpass.getpass("Enter your AWS SECRET_KEY")
# SESSION_TOKEN = getpass.getpass("Enter your AWS SESSION_TOKEN")
```

```python
import boto3
import json

print (f"ACCESS_KEY is set to {ACCESS_KEY[:4]}....")
print (f"SECRET_KEY is set to {SECRET_KEY[:4]}....")

# Setup bedrock
client = boto3.client(
    service_name="bedrock-runtime",
    region_name="us-east-1",
    aws_access_key_id=ACCESS_KEY, 
    aws_secret_access_key=SECRET_KEY
)


# Set the model ID, e.g., Claude 3 Haiku.
model_id = "anthropic.claude-3-haiku-20240307-v1:0"
```

```python

import requests
import html2text

# The URL
link='https://www.jspsych.org/latest/'
link='https://www.jspsych.org/latest/tutorials/rt-task/'
f = requests.get(link)

# URL content to plain text (HTML)
textHtml = f.text

# HTML text to MD text
h = html2text.HTML2Text()
textMd = h.handle(textHtml) 
textMd = textMd[:30000]
print (textMd)
```

```python



# Define the prompt for the model.
prompt = f"Here is a portion of a web page {link} converted to markdown:\n\n{textMd}\n\nCan you please summarize that web page for a neuroscience researcher? The summary should be limited to 3 or 4 sentences."

# Format the request payload using the model's native structure.
native_request = {
    "anthropic_version": "bedrock-2023-05-31",
    "max_tokens": 512,
    "temperature": 0.1,
    "messages": [
        {
            "role": "user",
            "content": [{"type": "text", "text": prompt}],
        }
    ],
}

# Convert the native request to JSON.
request = json.dumps(native_request)

try:
    # Invoke the model with the request.
    response = client.invoke_model(modelId=model_id, body=request)

except (ClientError, Exception) as e:
    print(f"ERROR: Can't invoke '{model_id}'. Reason: {e}")
    exit(1)

# Decode the response body.
model_response = json.loads(response["body"].read())

# Extract and print the response text.
response_text = model_response["content"][0]["text"]
print(response_text)

```
